```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('patient.csv')
data.head(2)
```

```python
data.keys()
```

```python
x = [1,2,3,4,5]
y = [5, 7, 4, 6, 8]
z = [6, 4, 5, 3, 7]

plt.figure(figsize = (10,5))
plt.scatter(x, y, label = 'nilai Y')
plt.plot(x, z, label = 'nilai Z', color = 'red')
plt.legend(loc='upper left')
plt.title("Contoh Grafik")
plt.xlabel("Nilai X")
plt.ylabel("Nilai Y")
plt.show()
```

```python
plt.figure(figsize = (15,10))
plt.hist(data['age'], 6, color = 'blue')
plt.show()
```

```python
plt.pie(data['gender'].value_counts(), colors = ['pink','orange'], labels = ['laki','Women'] ,startangle = 90, autopct = '%.1f%%')
plt.legend()
plt.show()
```
